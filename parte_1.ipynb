#### Punto 1.1 Búsqueda por Grilla

```python
def test_function(x, y):
    return 0.26*(x**2 + y**2 - 0.48*x*y)
```

```python
#Muestras que valos a evaluar en la busqueda en grilla
x_axis= np.arange(-10,10.1,0.1)
y_axis= np.arange(-10,10.1,0.1)
x,y = np.meshgrid(x_axis,y_axis)

z= test_function(x,y)

#Busqueda en grilla
f_opt, best_x, best_y = valor_optimo(z)

x_best = x[best_x][best_y]
y_best = y[best_x][best_y]
print('Punto optimo: x:',x_best,'y:',y_best)
print('Valor optimo:' ,f_opt)
```

```python
#Grafica
grafica(x,y,z,x_best,y_best)
```
